**SENTIMENT ANALYSIS**

```python
from textblob import TextBlob

def Get_Sentiment_polarity(text):
  if pd.isna(text) or text == '':
    return 0
  blob = TextBlob(text)
  return blob.sentiment.polarity

df['sentiment_polarity'] = df['review'].apply(Get_Sentiment_polarity)
df
```

```python
def Get_Sentiment_category(polarity):
  if polarity > 0.1:
    return 'positive'
  elif polarity < -0.1:
    return 'negative'
  else:
    return 'neutral'
df['sentiment_category'] = df['sentiment_polarity'].apply(Get_Sentiment_category)
df

```

```python
def Get_Sentiment_Subjectivity(text):
  if pd.isna(text) or text == '':
    return 0
  blob = TextBlob(text)
  return blob.sentiment.subjectivity
df['sentiment_subjectivity'] = df['review'].apply(Get_Sentiment_Subjectivity)
df
```

```python
import matplotlib.pyplot as plt
seniment_counts = df['sentiment_category'].value_counts()
colours = ['#2ecc71','#f39c12','#e74c3c']
plt.bar(x=seniment_counts.index,height=seniment_counts.values,color=colours)
plt.title('Sentiment Counts')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(data=df,x='sentiment_subjectivity',bins=10,alpha=0.7,color='#2ecc71',edgecolor='black')
plt.axvline(x=df['sentiment_subjectivity'].mean(),color='red',linestyle='dashed',linewidth=2,label=f"Mean: {df['sentiment_polarity'].mean():.3f}")
plt.legend()
plt.title('Sentiment Subjectivity Distribution')
plt.xlabel('Subjectivity')
plt.ylabel('Count')
plt.show()
#
```

**ANALYZING POSITIVE FEATURES**

```python
positive_reviews = df[df['sentiment_category'] == 'positive']
```

```python
import re
from typing import Counter

all_text = ' '.join(positive_reviews['review'].astype(str))
cleantext = re.sub(r'[^a-zA-Z\s]','',all_text)
words = cleantext.split()

words = [word for word in words if len(words) > 2]

phrases = []
for i in range(len(words) - 1):
  if len(words[i]) > 2 and len(words[i + 1]) > 2:
    phrases.append(f"{words[i]} {words[i + 1]}")

for i in range(len(words) - 2):
  if all(len(word) > 2 for word in words[i:i+3]):
    phrases.append(f"{words[i]} {words[i + 1]} {words[i + 2]}")
phrase_freq = Counter(phrases)
pos_phrase = phrase_freq.most_common(10)
```

```python
import matplotlib.pyplot as plt

phrases = [phrase for phrase, count in pos_phrase]
counts = [count for phrase, count in pos_phrase]

plt.figure(figsize=(10, 6))
plt.bar(phrases, counts, color='skyblue')
plt.xlabel('Phrases')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Positive Review Phrases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**ANALYZING NEGATIVE FEATURES**

```python
negative_reviews = df[df['sentiment_category'] == 'negative']

all_text_negative = ' '.join(negative_reviews['review'].astype(str))
cleantext_negative = re.sub(r'[^a-zA-Z\s]','',all_text_negative)
words_negative = cleantext_negative.split()
```

```python
words_negative = [word for word in words_negative if len(words_negative) > 2]

phrases_negative = []
for i in range(len(words_negative) - 1):
  if len(words_negative[i]) > 2 and len(words_negative[i + 1]) > 2:
    phrases_negative.append(f"{words_negative[i]} {words_negative[i + 1]}")

for i in range(len(words_negative) - 2):
  if all(len(word) > 2 for word in words_negative[i:i+3]):
    phrases_negative.append(f"{words_negative[i]} {words_negative[i + 1]} {words_negative[i + 2]}")

phrase_freq_negative = Counter(phrases_negative)
neg_phrase = phrase_freq_negative.most_common(10)
```

```python
phrases_plot_negative = [phrase for phrase, count in neg_phrase]
counts_plot_negative = [count for phrase, count in neg_phrase]

plt.figure(figsize=(10, 6))
plt.bar(phrases_plot_negative, counts_plot_negative, color='salmon')
plt.xlabel('Phrases')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Negative Review Phrases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Define some potential issue categories and their keywords
issue_categories = {
    'Performance/Bugs': ['lag', 'slow', 'crash', 'bug', 'glitch', 'error', 'broken', 'freezes'],
    'Cost/Pricing': ['price', 'cost', 'expensive', 'money', 'premium', 'subscription', 'upgrade', 'pay'],
    'Accuracy/Quality': ['wrong answer', 'incorrect', 'bad answer', 'not accurate', 'bad quality', 'misleading', 'false', 'nonsense', 'irrelevant'],
    'Features/Functionality': ['missing', 'lack of', 'no feature', 'cannot', 'can\'t', 'limited', 'unable', 'doesn\'t work', 'functionality'],
    'User Interface/Experience': ['hard to use', 'confusing', 'UI', 'UX', 'interface', 'design', 'clunky', 'difficult'],
    'Censorship/Restrictions': ['restrict', 'censor', 'blocked', 'limit', 'bias']
}

# Initialize a dictionary to store counts for each category
category_counts = {category: 0 for category in issue_categories}

# Convert all negative reviews to lowercase for case-insensitive matching
negative_reviews_lower = negative_reviews['review'].astype(str).str.lower()

# Iterate through each negative review
for review_text in negative_reviews_lower:
    for category, keywords in issue_categories.items():
        # Check if any keyword for the current category is present in the review
        if any(keyword in review_text for keyword in keywords):
            category_counts[category] += 1
            # Break to avoid double-counting if multiple keywords from the same category are in one review
            break

# Convert to a pandas Series for easier plotting
issue_category_series = pd.Series(category_counts)

display(issue_category_series.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt

# Plotting the results
plt.figure(figsize=(12, 7))
issue_category_series.sort_values(ascending=False).plot(kind='bar', color='darkred')
plt.title('Distribution of Issue Categories in Negative Reviews')
plt.xlabel('Issue Category')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**ANALYZING NEUTRAL FEATURES**

```python
neutral_subcategories_keywords = {
    'Implicitly Negative': [
        'not working', 'issue', 'problem', 'bug', 'glitch', 'error', 'crashes', 'slow', 'lag',
        'failed', 'broken', 'difficult', 'hard', 'unstable', 'useless', 'bad', 'disappointing',
        'worst', 'horrible', 'annoying', 'frustrating', 'fix this', 'needs improvement'
    ],
    'Questions': [
        'how to', 'can i', 'what about', 'where is', 'why not', 'is there', 'wondering',
        'question', 'help me', 'need to', 'do you', 'can you', 'ask about'
    ],
    'Feature Requests': [
        'add feature', 'wishlist', 'should have', 'need a', 'want to', 'suggest', 'improve',
        'update', 'include', 'could you add', 'expecting', 'future updates', 'more options'
    ]
}

print("Defined neutral_subcategories_keywords dictionary.")
```

```python
neutral_reviews = df[df['sentiment_category'] == 'neutral'].copy()
print(f"Extracted {len(neutral_reviews)} neutral reviews.")
```

```python
def assign_neutral_subcategory(review_text):
    review_text_lower = str(review_text).lower()
    for category, keywords in neutral_subcategories_keywords.items():
        if any(keyword in review_text_lower for keyword in keywords):
            return category
    return 'Truly Neutral'

neutral_reviews['neutral_subcategory'] = neutral_reviews['review'].apply(assign_neutral_subcategory)

neutral_subcategory_counts = neutral_reviews['neutral_subcategory'].value_counts()
display(neutral_subcategory_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
neutral_subcategory_counts.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Distribution of Neutral Review Subcategories')
plt.xlabel('Neutral Subcategory')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
monthly_reviews = df.resample('ME').size()
print(monthly_reviews.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
monthly_reviews.plot(kind='line', marker='o', color='skyblue')
plt.title('Monthly Review Counts')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_avg_ratings = df['ratings'].resample('ME').mean()

plt.figure(figsize=(12, 6))
monthly_avg_ratings.plot(kind='line', marker='o', color='lightcoral')
plt.title('Monthly Average Ratings')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
